```python
### 1. 
import pandas as pd 
import time  
start_time = time.time()
df = pd.read_csv('/Users/hkkkkk/Documents/GitHub/Python2/problem_sets/ps1/data/parking_tickets_one_percent.csv')
end_time = time.time()
elapsed_time = end_time - start_time

print(f"Time taken to read the file: {elapsed_time:.4f} seconds")
assert len(df) == 287458, f"Expected 287458 rows, but got {len(df)}"
print("Assertion passed: The number of rows is correct.")

### Time taken to read the file: 0.7781 seconds
### Assertion passed: The number of rows is correct.

### 2. 
import os 
file_path = '/Users/hkkkkk/Documents/GitHub/Python2/problem_sets/ps1/data/parking_tickets_one_percent.csv'
file_size_bytes = os.path.getsize(file_path)
file_size_megabytes = file_size_bytes / (1024 * 1024)

print(f"Size of the CSV file: {file_size_megabytes:.2f} MB")

### Size of the CSV file: 79.78 MB
  
### 3. 
print(df.head())
subset_df = df.head(500)
def is_column_ordered(column):
    return all(column[i] <= column[i + 1] for i in range(len(column) - 1))

column_name = 'ticket_number'  
if column_name in subset_df.columns:
    ordered = is_column_ordered(subset_df[column_name])
    print(f"The column '{column_name}' is ordered: {ordered}")
else:
    print(f"Column '{column_name}' not found in the DataFrame.")

column_name = 'issue_date'  
if column_name in subset_df.columns:
    ordered = is_column_ordered(subset_df[column_name])
    print(f"The column '{column_name}' is ordered: {ordered}")
else:
    print(f"Column '{column_name}' not found in the DataFrame.")
### The column 'issue_date' is ordered: True
### The data set is ordered by the issue date 
```

```python
### 1. 
df['issue_date'] = pd.to_datetime(df['issue_date'])
tickets_2017 = df[df['issue_date'].dt.year == 2017]
count_tickets_2017 = tickets_2017.shape[0]
print(f"Tickets issued in 2017 (sample): {count_tickets_2017}")
full_count_tickets_2017 = count_tickets_2017 * 100
print(f"Estimated tickets issued in 2017 (full dataset): {full_count_tickets_2017}")
### Tickets issued in 2017 (sample): 2236400
### According to the article, Chicago issues more than 3 million tickets, which is close to the result 

### 2. 
import altair as alt
violation_counts = df['violation_description'].value_counts()
top_20_violations = violation_counts.head(20)
print(top_20_violations)
```

```python
chart = alt.Chart(top_20_violations).mark_bar().encode(
    alt.X('violation_description:N', title='Violation Description', sort='-y'),
    alt.Y('frequency:Q', title='Frequency')
).properties(
    width=800,
    height=400
)

chart
```

```python
### 1. 
import pandas as pd

# Define the column information
column_info = [
    ('ticket_number', 'Nominal'),
    ('issue_date', 'Temporal'),
    ('violation_location', 'Nominal'),
    ('license_plate_number', 'Nominal'),
    ('license_plate_state', 'Nominal'),
    ('license_plate_type', 'Nominal'),
    ('zipcode', 'Nominal'),
    ('violation_code', 'Nominal'),
    ('violation_description', 'Nominal'),
    ('unit', 'Nominal'),
    ('unit_description', 'Nominal'),
    ('vehicle_make', 'Nominal'),
    ('fine_level1_amount', 'Quantitative'),
    ('fine_level2_amount', 'Quantitative'),
    ('current_amount_due', 'Quantitative'),
    ('total_payments', 'Quantitative'),
    ('ticket_queue', 'Nominal'),
    ('ticket_queue_date', 'Temporal'),
    ('notice_level', 'Nominal'),
    ('hearing_disposition', 'Nominal'),
    ('notice_number', 'Nominal'),
    ('officer', 'Nominal'),
    ('address', 'Nominal')
]
data_types = pd.DataFrame(column_info, columns=['Variable Name', 'Data Types'])

print(data_types.to_markdown(index=False))
```

```python
### 2.  
paid_tickets = df[df['ticket_queue'] == 'Paid']
vehicle_make_fraction = df.groupby('vehicle_make').agg(
    total_tickets=('ticket_number', 'count'),
    paid_tickets=('ticket_queue', lambda x: (x == 'Paid').sum())
)

vehicle_make_fraction['fraction_paid'] = vehicle_make_fraction['paid_tickets'] / vehicle_make_fraction['total_tickets']

vehicle_make_fraction.reset_index(inplace=True)
```

```python
### 3. 
import altair as alt
chart = alt.Chart(vehicle_make_fraction).mark_bar().encode(
    x=alt.X('vehicle_make:N', title='Vehicle Make', sort='-y'),
    y=alt.Y('fraction_paid:Q', title='Fraction of Paid Tickets')
).properties(
    title='Fraction of Paid Tickets by Vehicle Make',
    width=600,
    height=400
).configure_axisX(
    labelAngle=-45  
)
chart
```

```python
### 4. 
df['issue_date'] = pd.to_datetime(df['issue_date'])
df['month'] = df['issue_date'].dt.month
df['day'] = df['issue_date'].dt.day

tickets_by_month_day = df.groupby(['month', 'day']).size().reset_index(name='ticket_count')

heatmap = alt.Chart(tickets_by_month_day).mark_rect().encode(
    x=alt.X('day:O', title='Day of Month'),
    y=alt.Y('month:O', title='Month'),
    color=alt.Color('ticket_count:Q', scale=alt.Scale(scheme='blues'), title='Number of Tickets Issued'),
).properties(
    title='Number of Tickets Issued by Month and Day',
    width=600,
    height=400
).configure_axisX(
    labelAngle=0
)
```
